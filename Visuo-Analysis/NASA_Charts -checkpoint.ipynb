```python
years = [1999, 2003, 2008, 2012, 2016]
meteorite_counts = [173, 120, 98, 85, 71]

plt.bar(years, meteorite_counts)
plt.xlabel('Year')
plt.ylabel('Number of Meteorite Falls')
plt.title('Meteorite Falls by Year')

plt.show()
```

```python
# Define the range of meteorite masses
mass_range = ['21g - 100g', '101g - 1,000g', '1,001g - 10,000g', '10,001g - 100,000g']

# Define the corresponding frequency of meteorites in each mass range
frequency = [150, 75, 30, 10]

# Create the area chart
plt.fill_between(mass_range, frequency, color='lightblue', alpha=0.5)

# Add labels and title
plt.xlabel('Mass Range')
plt.ylabel('Frequency')
plt.title('Mass of Meteorites')

# Display the area chart
plt.show()
```

```python
# Define the years and corresponding number of meteorite falls
years = [2000, 2005, 2010, 2015, 2020]
falls = [10, 15, 8, 12, 18]

# Create the line graph
plt.plot(years, falls, marker='o')

# Add labels and title
plt.xlabel('Years')
plt.ylabel('Number of Meteorite Falls')
plt.title('Meteorite Falls Over Time')

# Display the line graph
plt.show()
```
